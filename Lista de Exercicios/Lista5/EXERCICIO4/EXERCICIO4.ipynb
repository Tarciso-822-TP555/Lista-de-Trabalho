```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.datasets import make_circles
from scipy.special import expit
import seaborn as sns
from matplotlib.colors import ListedColormap
```

```python
np.random.seed(42)
```

```python
x1 = np.linspace(0,10,100).reshape(100,1)
y = 5 - x1
h = expit(y)
```

```python
# Create color maps
cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

plt.figure(figsize=(7,7))
# Plot the decision boundary. For that, we will assign a color to each point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = x1.min() - 1, x1.max() + 1
xx = np.arange(x_min, x_max, 0.02)

xx = xx.reshape(len(xx), 1)

yy = 5 - xx
Z = expit(yy)

Y = []
for i in range(0,len(h)):
    if(h[i] >= 0.5):
        Y.append(True)
    else:
        Y.append(False)

plt.axhspan(ymin=0.0, ymax=1.0, xmin=0.0, xmax=0.5, facecolor='blue', alpha=0.5)
plt.axhspan(ymin=0.0, ymax=1.0, xmin=0.5, xmax=1.0, facecolor='yellow', alpha=0.5)

# Plot also the training points.
plt.scatter(x1.ravel(), h.ravel(), c=Y, cmap=cmap_bold, edgecolor='k', s=20)

plt.xlabel('$x$', fontsize=14)
plt.ylabel('$h$', fontsize=14)
plt.text(1.5, 0.5, 'Classe +', fontsize=14)
plt.text(6.5, 0.5, 'Classe -', fontsize=14)
plt.grid()

plt.show()
```
