```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = np.random.randn(1000000)
# histograma (pdf)
plt.subplot(1, 2, 1)
plt.title('PDF')
plt.hist(data , bins=100, density=True , color='b')
# CDF empirica
plt.subplot(1, 2, 2)
plt.title('CDF')
plt.hist(data , bins=100, density=True , color='g', cumulative=True)
plt.savefig('Exercise 2 - Histogram.png') # salva figura em arquivo
```
