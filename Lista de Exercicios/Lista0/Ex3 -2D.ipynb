```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

#Number of Elements
M = 1000

x = np.random.rand(M) # Uniform Distribution
w = np.random.randn(M) # Gaussian Distribution

y = 1.2 + 2.3*x + 10*w

fig1, axis = plt.subplots()
#axis.plot(x, y)

#Plot points
plt.scatter(x, y, color = "red")

axis.set(xlabel='x', ylabel='y')
plt.grid()
plt.show()
```
